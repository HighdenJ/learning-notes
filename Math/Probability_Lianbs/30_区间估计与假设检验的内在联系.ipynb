### 假设检验的基本思想

在**区间估计**中，我们的目标是“**估计参数**”：给出一个区间，较大概率包含真实参数。

在**假设检验**中，目标变成“**判断一个关于参数的说法对不对**”：  
- 不再直接给出参数的估计值；  
- 而是先提出一个“假设”（比如 $\mu = \mu_0$），再用样本数据来“检验”它是否合理。

---

### 核心思想：小概率事件在一次试验中几乎不发生

假设检验的逻辑可以概括为：

> 1. 先假定一个说法为真（原假设 $H_0$ 成立）；  
> 2. 在这个假定下，算出“我们现在这组样本”出现的概率情况；  
> 3. 如果这组样本属于“小概率事件”（概率很小），就认为：  
>    “在 $H_0$ 为真时，这样离谱的样本几乎不可能出现”，  
>    因此**拒绝 $H_0$**。

这就是“**小概率事件在一次试验中几乎不发生**”+“**反证法**”的结合。

---

### 假设的两种类型：$H_0$ 与 $H_1$

- **原假设（零假设） $H_0$**：  
  通常表示“现状”、“没有差别”、“没有效果”的说法，例如  
  - $H_0:\ \mu = \mu_0$  
  - $H_0:\ p = p_0$

- **备择假设 $H_1$**：  
  表示与 $H_0$ 相对立的说法，例如  
  - $H_1:\ \mu \ne \mu_0$（双侧检验）  
  - $H_1:\ \mu > \mu_0$ 或 $\mu < \mu_0$（单侧检验）

检验的结果永远只有两种：  
- “拒绝 $H_0$”（倾向于 $H_1$）  
- “不拒绝 $H_0$”（暂时没有足够证据否定 $H_0$，但不等于证明它绝对正确）

---

### 假设检验的分类

1. **参数检验**  
   - 已知总体分布的形式（如正态），但参数未知；  
   - 检验对象是**参数**（如均值 $\mu$、方差 $\sigma^2$、比例 $p$ 等）；  
   - 例如：检验 $N(\mu,\sigma^2)$ 中的 $\mu$ 是否等于某个值。

2. **非参数检验**  
   - 不具体假定总体分布的形式；  
   - 检验对象是**分布本身**或“是否来自同一分布”；  
   - 例如：符号检验、秩和检验、Kolmogorov–Smirnov 检验等。

---

### 检验的基本流程（结构化版）

设样本为 $X_1,\dots,X_n$。

1. **提出假设**  
   - 写出原假设 $H_0$，备择假设 $H_1$。  
   - 例如：$H_0:\ \mu = \mu_0$，$H_1:\ \mu \ne \mu_0$。

2. **选择检验统计量**  
   - 根据模型和参数，构造一个关于样本的函数  
     $$
     T = T(X_1,\dots,X_n)
     $$
     称为**检验统计量**；  
   - 在 $H_0$ 成立的前提下，知道（或近似知道）$T$ 的分布。

3. **根据显著性水平确定“小概率区间”（拒绝域）**  
   - 选定显著性水平 $\alpha$（例如 $\alpha = 0.05$）；  
   - 在 $H_0$ 成立的前提下，求出检验统计量 $T$ 的分布分界点（临界值）；  
   - 把“概率为 $\alpha$ 的那部分极端区域”定义为**拒绝域**（小概率区间），其余为**接受域**（大概率区间）。

4. **代入样本，计算统计量并作出判断**  
   - 用观测到的样本值算出 $T_{\text{obs}}$；  
   - 若 $T_{\text{obs}}$ 落在拒绝域（小概率区间），就认为“在 $H_0$ 成立时几乎不可能出现这样的样本”，**拒绝 $H_0$**；  
   - 否则，**不拒绝 $H_0$**。

---

### 显著性水平与两类错误（只需有个印象）

在实际中：

- **第一类错误（Type I error，弃真）**：  
  $H_0$ 为真，却被错误地拒绝。  
  - 概率记为 $\alpha$，就是事先选定的**显著性水平**。

- **第二类错误（Type II error，取伪）**：  
  $H_0$ 为假，却没有被拒绝。  
  - 概率通常记为 $\beta$，$1-\beta$ 称为**检验的功效**（power）。

常见做法：先固定一个可接受的第一类错误水平 $\alpha$（例如 0.05），在此基础上尽量使检验“有力”（功效大）。

---

### 与区间估计的联系

很多情况下，**假设检验**和**区间估计**是等价的：

- 比如检验 $H_0:\ \mu = \mu_0$，显著性水平为 $\alpha$；  
- 若 $\mu_0$ 落在 $\mu$ 的 $1-\alpha$ 置信区间内，则“不拒绝 $H_0$”；  
- 若 $\mu_0$ 不在该置信区间内，则“拒绝 $H_0$”。

因此，可以把假设检验看成是：  
> “问：某个给定的参数值，是否被样本构造出的置信区间‘支持’？”

### 练习题 1：正态总体均值的假设检验（σ 已知）

某日某工厂生产的某种产品，包装袋上标称净含量为 **500 g**。  
设每袋产品的实际净含量为随机变量 $X$，服从正态分布
$$
X \sim N(\mu, 2^2)\ \text{(g)},
$$
其中均值 $\mu$ 未知，但方差 $\sigma^2 = 2^2$ 已知。

为检验该生产工艺是否仍然达标，从当天的产品中**随机抽取 8 袋**，测得净含量如下（单位：g）：
$$
503,\ 497,\ 502,\ 506,\ 495,\ 490,\ 497,\ 500.
$$

在显著性水平 $\alpha = 0.05$ 下，能否认为该产品的平均净含量仍为 **500 g**？

---

### 解答

#### 第 1 步：提出假设

我们关心的是平均净含量是否仍为 500 g，因此设

- 原假设：
  $$
  H_0:\ \mu = 500;
  $$
- 备择假设（双侧）：
  $$ 
  H_1:\ \mu \ne 500.
  $$

这是一个 **正态总体均值检验（方差已知）** 的双侧检验。

---

#### 第 2 步：选取检验统计量及其分布

样本容量为 $n=8$，样本均值为
$$
\bar X = \frac{1}{8}(503+497+502+506+495+490+497+500)=498.75\ \text{(g)}.
$$

在 $H_0:\mu=500$ 成立且方差 $\sigma^2=2^2$ 已知的条件下，检验统计量取为
$$
Z = \frac{\bar X - \mu_0}{\sigma/\sqrt{n}},
$$
其中 $\mu_0 = 500$。此时
$$
Z \sim N(0,1)\quad (\text{在 } H_0 \text{ 下}).
$$

---

#### 第 3 步：给定显著性水平，确定拒绝域

显著性水平 $\alpha = 0.05$，双侧检验，对应标准正态分布的临界值为
$$
z_{0.025} \approx 1.96.
$$

拒绝域：
$$
\{|Z| > 1.96\}.
$$

---

#### 第 4 步：计算样本的检验统计量

代入样本值：
$$
Z_{\text{obs}}
= \frac{\bar X - 500}{2/\sqrt{8}}
= \frac{498.75 - 500}{2/\sqrt{8}}
= \frac{-1.25}{2/\sqrt{8}}
= \frac{-1.25}{0.7071}
\approx -1.77.
$$

因此
$$
|Z_{\text{obs}}| \approx 1.77.
$$

---

#### 第 5 步：作出统计结论

比较：
$$
|Z_{\text{obs}}| \approx 1.77 < 1.96 = z_{0.025},
$$
因此 **$Z_{\text{obs}}$ 未落入拒绝域**，在显著性水平 $\alpha = 0.05$ 下，**不能拒绝原假设 $H_0$**。

也可以看 $p$ 值：
$$
p = 2P(Z > |Z_{\text{obs}}|)
\approx 2(1-\Phi(1.77))
\approx 0.077 > 0.05,
$$
同样说明不能拒绝 $H_0$。

---

### 结论（文字表述）

在显著性水平 $0.05$ 下，根据这 8 袋产品的净含量数据，**没有足够的统计证据认为平均净含量已经偏离 500 g**。  
因此，我们 **不能拒绝“平均净含量为 500 g”这一假设**，也就是可以认为该生产工艺在目前的检验标准下仍然是“合格”的。

### 练习题 2：正态总体均值的假设检验（$\sigma$ 未知）

某工厂生产的一种轴承，其设计要求**平均寿命为 2000 小时**。  
为了检验目前生产是否仍然达标，从当天产出的轴承中随机抽取了 10 个，测得寿命（单位：小时）如下：

$$
1980,\ 2015,\ 2030,\ 1995,\ 2050,\ 1970,\ 2005,\ 2020,\ 1965,\ 2040.
$$

假设轴承寿命 $X$ 服从正态分布 $N(\mu,\sigma^2)$，其中 $\mu,\sigma^2$ 均未知。

请在显著性水平 $\alpha = 0.05$ 下，对“平均寿命是否仍为 2000 小时”进行检验。



### 解答

#### 1. 原假设和备择假设

- 原假设（零假设）  
  $$
  H_0:\ \mu = 2000
  $$
- 备择假设（双侧检验）  
  $$
  H_1:\ \mu \ne 2000
  $$

---

#### 2. 检验统计量及其分布

总体 $X \sim N(\mu,\sigma^2)$，但 $\sigma$ 未知，用样本标准差 $S$ 代替。

检验统计量取为  
$$
T = \frac{\bar X - \mu_0}{S/\sqrt{n}},
$$
其中 $\mu_0 = 2000$，$n=10$。

*备注：上式$T$服从$t$分布的证明见`23_24_统计量的构造.ipynb`*

在 $H_0$ 成立时，有  
$$
T \sim t(n-1) = t(9).
$$


---

#### 3. 计算样本统计量

样本数据：
$$
1980,\ 2015,\ 2030,\ 1995,\ 2050,\ 1970,\ 2005,\ 2020,\ 1965,\ 2040.
$$

- 样本均值：
  $$
  \bar X = \frac{1}{10}\sum_{i=1}^{10} X_i
  = \frac{20070}{10}
  = 2007.
  $$

- 样本标准差：
  $$
  S = \sqrt{\frac{1}{n-1}\sum_{i=1}^{10}(X_i - \bar X)^2}
  \approx 29.27.
  $$

- 检验统计量的观测值：
  $$
  t_{\text{obs}}
  = \frac{\bar X - \mu_0}{S/\sqrt{n}}
  = \frac{2007 - 2000}{29.27/\sqrt{10}}
  \approx \frac{7}{9.25}
  \approx 0.76.
  $$

---

#### 4. 拒绝域

显著性水平 $\alpha = 0.05$，双侧检验，自由度 $n-1 = 9$。  
查 $t$ 分布表得
$$
t_{0.025}(9) \approx 2.262.
$$

拒绝域：
$$
\{|T| > 2.262\}.
$$

---

#### 5. 判决与结论

观察值：
$$
|t_{\text{obs}}| \approx 0.76 < 2.262,
$$
**未落入拒绝域**，因此在显著性水平 $\alpha = 0.05$ 下，**不能拒绝原假设 $H_0$**。

> 通俗结论：  
> 根据这 10 个样本的数据，没有足够的统计证据说明轴承的平均寿命已经偏离 2000 小时，可以认为“平均寿命仍为 2000 小时”的说法是合理的。

### 练习题 3：给定拒绝域时的第一类错误概率

某产品的**标准长度**为 68 mm。设产品长度 $X$ 服从正态分布
$$
X \sim N(\mu, 3.6^2)\ \text{(mm)}.
$$

要检验该产品的平均长度是否仍为标准值 68 mm，提出假设：

- 原假设（零假设）  
  $$
  H_0:\ \mu = 68;
  $$
- 备择假设  
  $$
  H_1:\ \mu \ne 68.
  $$

抽取容量为 $n$ 的样本，记样本均值为 $\bar X$。  
给定如下**判别规则**：

- 若 $|\bar X - 68| > 1$，则 **拒绝** $H_0$；  
- 若 $|\bar X - 68| \le 1$，则 **接受**（不拒绝）$H_0$。

问：

1. 当样本容量为 $n = 36$ 时，犯第一类错误的概率 $\alpha$ 为多少？  
2. 当样本容量为 $n = 64$ 时，犯第一类错误的概率 $\alpha$ 为多少？

---

### 解答思路

**第一类错误**：在 $H_0$ 为真（$\mu=68$）的情况下，错误地拒绝 $H_0$ 的事件。  
本题中，
$$
\alpha
= P(\text{拒绝 }H_0 \mid H_0\text{ 为真})
= P\big(|\bar X - 68| > 1 \,\big|\, \mu = 68\big).
$$

在 $H_0$ 成立且方差已知为 $3.6^2$ 时，样本均值的分布为
$$
\bar X \sim N\!\left(68,\ \frac{3.6^2}{n}\right).
$$

令
$$
Z = \frac{\bar X - 68}{3.6/\sqrt{n}} \sim N(0,1).
$$

于是
$$
\alpha
= P\left(|\bar X - 68| > 1\right)
= P\left(\left|\frac{\bar X - 68}{3.6/\sqrt{n}}\right| > \frac{1}{3.6/\sqrt{n}}\right)
= P\big(|Z| > c_n\big),
$$
其中
$$
c_n = \frac{1}{3.6/\sqrt{n}} = \frac{\sqrt{n}}{3.6}.
$$

因此
$$
\alpha = 2\big[1 - \Phi(c_n)\big],
$$
$\Phi(\cdot)$ 为标准正态分布函数。

---

### 第 1 问：$n=36$

此时
$$
c_{36} = \frac{\sqrt{36}}{3.6} = \frac{6}{3.6} = \frac{5}{3} \approx 1.67.
$$

故
$$
\alpha
= 2\big[1 - \Phi(5/3)\big].
$$

若查表 $\Phi(1.67)\approx 0.952$（近似），则
$$
\alpha \approx 2(1 - 0.952) \approx 2 \times 0.048 \approx 0.096.
$$

所以，
$$
\boxed{\alpha \approx 0.096\quad (n=36).}
$$

---

### 第 2 问：$n=64$

此时
$$
c_{64} = \frac{\sqrt{64}}{3.6} = \frac{8}{3.6} \approx 2.22.
$$

故
$$
\alpha
= 2\big[1 - \Phi(8/3.6)\big]
\approx 2\big[1 - \Phi(2.22)\big].
$$

若查表 $\Phi(2.22)\approx 0.987$（近似），则
$$
\alpha \approx 2(1 - 0.987) \approx 2 \times 0.013 \approx 0.026.
$$

所以，
$$
\boxed{\alpha \approx 0.026\quad (n=64).}
$$

---

### 小结

- 同一拒绝规则下，增加样本容量 $n$ 会使样本均值的波动变小（方差 $\sigma^2/n$ 变小），  
- 相同的“阈值” $|\bar X -68|>1$ 在标准化后对应更大的 $c_n$，  
- 从而第一类错误概率 $\alpha = 2[1 - \Phi(c_n)]$ 随着 $n$ 增大而减小。

### p-value 及其与假设检验的关系

### 1. 在传统假设检验中的位置

传统（Neyman–Pearson）假设检验流程里，关键是：

1. 给定显著性水平 $\alpha$（比如 $0.05$）；  
2. 根据 $H_0$ 和 $\alpha$，事先确定**拒绝域**（临界值、分界点）；  
3. 看统计量是否落入拒绝域，做“拒绝 / 不拒绝”判断。

**p-value** 是在这个框架基础上的一个“**附加衡量指标**”，不再只给出“拒绝 / 不拒绝”，而是给出一个“**有多不利于 $H_0$**”的程度，或者说用来**衡量“样本数据对原假设 $H_0$ 的不利程度”**。

---

### 2. p-value 的定义

假设检验中：

- 先假定 $H_0$ 成立；
- 计算出一个检验统计量的观测值 $t_{\text{obs}}$；
- **p-value** 定义为：

> 在 $H_0$ 为真时，抽到“像现在这么极端或更极端”的样本的概率。

符号上（以**右侧检验**为例）：
$$
p\text{-value}
= P_\theta(T \ge t_{\text{obs}} \mid H_0).
$$

- 双侧检验通常是：  
  $$
  p\text{-value}
  = 2P_\theta(T \ge |t_{\text{obs}}| \mid H_0).
  $$

---

### 3. p-value 和显著性水平 $\alpha$ 的关系

**决策规则：**

- 若 $p \le \alpha$：拒绝 $H_0$；  
- 若 $p > \alpha$：不拒绝 $H_0$。

所以你可以这样理解：

> 传统方法：**先**选 $\alpha$，再看统计量。  
> p-value 方法：**先**算 p-value，再根据不同的 $\alpha$ 自由决定要不要拒绝。

p-value 越小，意味着：

- 在 $H_0$ 为真的情况下，出现当前这种样本的概率越小；
- 数据对 $H_0$ 越“不友好”，越倾向于拒绝 $H_0$。

---

### 4. 一个例子（标准流程）

假设检验：

- $H_0: \mu = \mu_0$  
- $H_1: \mu \ne \mu_0$  
- 检验统计量（$\sigma$ 未知）：
  $$
  T = \frac{\bar X - \mu_0}{S/\sqrt{n}} \sim t(n-1)\quad (H_0\text{ 下}).
  $$
- 得到观测值 $t_{\text{obs}}$。

双侧检验的 p-value：
$$
p = 2P\big(T \ge |t_{\text{obs}}|\mid H_0\big)
= 2\big(1 - F_t(|t_{\text{obs}}|)\big),
$$
其中 $F_t$ 是自由度为 $n-1$ 的 $t$ 分布的 CDF。

---

### 5. 常见误解（非常重要）

1. **p-value 不是** “$H_0$ 为真的概率”。  
   - 正确理解：是在假定 $H_0$ 已经“当真成立”的前提下，数据有多“离谱”。

2. **p-value 大不代表 $H_0$ 就是对的**。  
   - 只代表“数据不足以推翻 $H_0$”；  
   - 样本太少、检验能力太弱时，p-value 很可能大，但这不等于证明 $H_0$ 正确。

3. “p-value = 0.03” 不是说“有 97% 的把握 $H_1$ 对”。  
   - 只是说：在 $H_0$ 真时，出现目前或更极端数据的概率是 3%。

---

### 6. 可以把 p-value 当什么？

你可以把 p-value 当成：

> 一个**衡量证据强度的刻度**：  
> - 小于 0.01：对 $H_0$ 很不利，证据很强；  
> - 大约在 0.01～0.05：证据比较强；  
> - 在 0.05～0.10：证据偏弱；  
> - 大于 0.10：数据与 $H_0$ 很相容，暂时看不出问题。

（界线只是经验说法，不是硬规则。）

---

### 7. 小结

- p-value 是在假设检验框架下引入的概念；  
- 它描述：**“在假定 $H_0$ 为真时，数据有多极端”**；  
- 计算方式依赖于：检验统计量的分布 + 备择假设是单侧还是双侧；  
- 决策规则：$p \le \alpha$ 拒绝 $H_0$，否则不拒绝；  
- p-value **不是** $H_0$ 为真的概率，也不是 $H_1$ 为真的概率。

掌握了 p-value，再看各种检验（$Z$ 检验、$t$ 检验、$\chi^2$ 检验、F 检验）就会发现：  
> 不管检验的名字怎么变，套路都是：先算统计量 → 再算 p-value → 再根据 $\alpha$ 做决策。

### 练习题 4：使用 p-value 进行假设检验（均值，σ 已知）

某电子产品的电池标称平均寿命为 100 小时，历史数据表明寿命近似服从正态分布  
$$
X \sim N(\mu, \sigma^2), \quad \sigma = 10\ \text{(小时)}.
$$

现在从新生产的一批电池中随机抽取了 $n=25$ 个样本，测得样本平均寿命为  
$$
\bar x = 104.5\ \text{小时}.
$$

我们想检验：**这批电池的平均寿命是否已经发生了变化**。

---

### 解答

#### 1. 假设

我们关注“平均寿命是否从 100 发生变化”，因此：

- 原假设  
  $$
  H_0: \mu = 100;
  $$
- 备择假设（双侧）  
  $$
  H_1: \mu \ne 100.
  $$

---

#### 2. 检验统计量及其分布

已知 $\sigma = 10$，样本容量 $n = 25$，样本均值为 $\bar X$。  
经典的 $Z$ 检验统计量为：

$$
Z = \frac{\bar X - \mu_0}{\sigma / \sqrt{n}},
$$

其中 $\mu_0 = 100$。  
在 $H_0$ 成立时，
$$
Z \sim N(0,1).
$$

---

#### 3. 观测到的统计量

将样本均值 $\bar x = 104.5$ 代入：

$$
z_{\text{obs}}
= \frac{\bar x - \mu_0}{\sigma / \sqrt{n}}
= \frac{104.5 - 100}{10 / \sqrt{25}}
= \frac{4.5}{10/5}
= \frac{4.5}{2}
= 2.25.
$$

---

#### 4. 计算 p-value（双侧）

双侧检验的 p-value 为：

$$
p = 2P(Z \ge |z_{\text{obs}}|)
= 2\big(1 - \Phi(2.25)\big),
$$

其中 $\Phi$ 为标准正态分布函数。

查表可得 $\Phi(2.25) \approx 0.9878$，于是

$$
p \approx 2(1 - 0.9878) = 2 \times 0.0122 \approx 0.0244.
$$

---

#### 5. 基于 p-value 的结论

在显著性水平 $\alpha = 0.05$ 下比较：

- $p \approx 0.0244 < 0.05$

因此：

> 在显著性水平 $0.05$ 下，**拒绝原假设 $H_0$**。

用通俗话说：

> 如果平均寿命真的还是 100 小时，那么像现在这样（或更极端）的样本均值出现的概率只有大约 2.4%。  
> 这个概率相当小，因此我们有理由认为“平均寿命已经发生了变化”。

---

### 6. Python 可视化：标准正态分布与 p-value 区域

- 在 $z = \pm |z_{\text{obs}}|$ 两侧的阴影区域（对应双侧 p-value）；
- 标出 $z_{\text{obs}}$ 和 p-value 的大致数值。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

alpha = 0.05
z_obs = 2.25                       # observed test statistic
p_value = 2 * (1 - norm.cdf(abs(z_obs)))
z_crit = norm.ppf(1 - alpha / 2)   # ≈ 1.96 for alpha=0.05

x = np.linspace(-4, 4, 400)
y = norm.pdf(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y)

# --------- Shade alpha region (|Z| > z_crit) ----------
x_alpha_left = np.linspace(-4, -z_crit, 200)
x_alpha_right = np.linspace(z_crit, 4, 200)

plt.fill_between(x_alpha_left, norm.pdf(x_alpha_left), alpha=0.15)
plt.fill_between(x_alpha_right, norm.pdf(x_alpha_right), alpha=0.15)

# Critical lines for alpha
plt.axvline(-z_crit, linestyle='-.')
plt.axvline(z_crit, linestyle='-.')

# --------- Shade p-value region (|Z| > |z_obs|) ----------
x_p_left = np.linspace(-4, -abs(z_obs), 200)
x_p_right = np.linspace(abs(z_obs), 4, 200)

plt.fill_between(x_p_left, norm.pdf(x_p_left), alpha=0.3)
plt.fill_between(x_p_right, norm.pdf(x_p_right), alpha=0.3)

# Lines for observed z
plt.axvline(-abs(z_obs), linestyle='--')
plt.axvline(abs(z_obs), linestyle='--')

plt.title("Standard normal: p-value region and alpha=0.05 critical region")
plt.xlabel("z")
plt.ylabel("density")

# Annotate critical values
plt.text(-z_crit+1.45, 0.01, rf'$-z_{{crit}}={-z_crit:.2f}$', ha="right", va="bottom", fontsize=11)
plt.text(z_crit-1.1, 0.01, rf'$z_{{crit}}={z_crit:.2f}$', ha="left", va="bottom", fontsize=11)

# Annotate observed z
plt.text(-abs(z_obs)-0.1, 0.04, rf'$-z_{{obs}}={-abs(z_obs):.2f}$', ha="right", va="bottom", fontsize=11)
plt.text(abs(z_obs)+0.1, 0.04, rf'$z_{{obs}}={abs(z_obs):.2f}$', ha="left", va="bottom", fontsize=11)

# Annotate p-value
plt.text(0, 0.2, rf'$p-\text{{value}} = {p_value:.4f}$', ha="center", va="center", fontsize=12, color='red')


plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
