### 点估计的选择与评价指标

回顾两种常见的**构造估计量**的方法：

- **矩估计**：用样本矩逼近总体矩  
  - 用样本均值 $\bar X$ 估计总体期望 $E[X]=\mu$  
  - 用样本方差 $S^2$ 估计总体方差 $D[X]=\sigma^2$
- **极大似然估计**：在所有可能的参数取值中，选取使“已经观察到的这组样本”出现的概率（或密度）最大的参数值

不同方法得到的估计量可能不一样，需要评判估计的好坏。用到下面三类性质（评价指标）。

---

### 1. 无偏性（unbiasedness）

设参数为 $\theta$，估计量为 $\hat\theta=\hat\theta(X_1,\dots,X_n)$。

**定义：**

如果
$$
E[\hat\theta] = \theta,
$$
则称 $\hat\theta$ 是 $\theta$ 的**无偏估计量**，或称 $\hat\theta$ 具有**无偏性**。

- 直观：在重复抽样、重复估计的长期平均意义下，估计值既不偏高也不偏低。

**例子：**

- 对于 $X_1,\dots,X_n$ 独立同分布且 $E[X_i]=\mu$，
  $$
  E[\bar X] = \mu \quad\Rightarrow\quad \bar X \text{ 是 } \mu \text{ 的无偏估计};
  $$
- 在常见定义下
  $$
  S^2 = \frac{1}{n-1}\sum_{i=1}^n (X_i - \bar X)^2,
  $$
  满足
  $$
  E[S^2] = \sigma^2 \quad\Rightarrow\quad S^2 \text{ 是 } \sigma^2 \text{ 的无偏估计}.
  $$

---

### 2. 有效性（efficiency）

当有**多个无偏估计量**时，需要比较它们的“波动大小”。

**定义（同样估计同一个参数 $\theta$）：**

若 $\hat\theta_1,\hat\theta_2$ 都是 $\theta$ 的无偏估计量，并且
$$
D[\hat\theta_1] < D[\hat\theta_2],
$$
则称 $\hat\theta_1$ 比 $\hat\theta_2$ **更有效**（更“集中”、更“稳定”）。

- 直观：在同样不偏的前提下，方差越小，估计越“准”。

有时候也会使用**均方误差**（MSE）来综合考虑偏差和方差：
$$
\mathrm{MSE}(\hat\theta) = E[(\hat\theta-\theta)^2]
= \underbrace{D(\hat\theta)}_{\text{方差}} + \underbrace{\big(E[\hat\theta]-\theta\big)^2}_{\text{偏差}^2}.
$$

---

### 3. 相合性（consistency）

无偏和有效关注的是“固定样本量 $n$ 下”的表现，而**相合性**关注的是“样本越来越多时，估计是否越来越接近真参数”。

**定义：**

如果对任意 $\varepsilon>0$，
$$
\lim_{n\to\infty}P\big(|\hat\theta_n - \theta| < \varepsilon\big) = 1,
$$
则称估计量序列 $\hat\theta_n$ 是 $\theta$ 的**相合估计量**，或称它具有**相合性**。

- 直观：样本量 $n$ 越大，$\hat\theta_n$ 越“紧紧贴近”真参数 $\theta$，偏离真值的概率越来越小。

**例子：**

在独立同分布的条件下，样本均值满足大数定律：
$$
\bar X_n \xrightarrow{P} \mu,
$$
因此 $\bar X_n$ 是 $\mu$ 的相合估计量。

---

### 小结：如何评价和选择估计量？

给定一个估计量 $\hat\theta$，可以按以下思路评判：

1. **先看无偏性**：  
   - 是否满足 $E[\hat\theta]=\theta$？  
   - 若不无偏，偏差是否能接受（用 MSE 衡量）？
2. **再比较有效性**（在若干无偏估计量之间）：  
   - 方差（或 MSE）谁更小？
3. **再看相合性**：  
   - 样本量增大时，$\hat\theta$ 是否稳定收敛到真值？  

在具体的分布和模型中，可以把矩估计、极大似然估计等方法得到的不同估计量放在一起，用这三类性质进行比较和选择。

### 练习题 1：根据评价指标的性质求解
> 设总体 $X \sim N(\mu,\sigma^2)$，从中抽取简单随机样本  
> $$
> X_1,X_2,\dots,X_n.
> $$
> 考虑统计量  
> $$
> \hat\sigma^2 = C \sum_{i=1}^{n-1} (X_{i+1}-X_i)^2.
> $$
> 求常数 $C$，使得 $\hat\sigma^2$ 是 $\sigma^2$ 的无偏估计量。

---

### 解答

要求 $\hat\sigma^2$ 是 $\sigma^2$ 的无偏估计量，即
$$
E[\hat\sigma^2] = \sigma^2.
$$

把定义代入：
$$
E[\hat\sigma^2]
= E\left[ C \sum_{i=1}^{n-1} (X_{i+1}-X_i)^2 \right]
= C \sum_{i=1}^{n-1} E\big[(X_{i+1}-X_i)^2\big].
$$

#### 第 1 步：计算单个差值的二阶矩

因为 $X_i$ 独立同分布 $N(\mu,\sigma^2)$，有  

- $E[X_i] = \mu$，$E[X_{i+1}] = \mu$；  
- $D(X_i) = D(X_{i+1}) = \sigma^2$；  
- $X_i$ 与 $X_{i+1}$ 独立。

令
$$
Y_i = X_{i+1}-X_i.
$$

则
$$
E[Y_i] = E[X_{i+1}] - E[X_i] = \mu - \mu = 0,
$$
方差为
$$
\begin{aligned}
D(Y_i)
&= D(X_{i+1}-X_i) \\
&= D(X_{i+1}) + D(X_i) \quad (\text{独立，协方差为 0}) \\
&= \sigma^2 + \sigma^2 \\
&= 2\sigma^2.
\end{aligned}
$$

又因为 $E[Y_i] = 0$，所以
$$
E[Y_i^2] = D(Y_i) = 2\sigma^2,
$$
即
$$
E\big[(X_{i+1}-X_i)^2\big] = 2\sigma^2.
$$

#### 第 2 步：代回求 $E[\hat\sigma^2]$

$$
\begin{aligned}
E[\hat\sigma^2]
&= C \sum_{i=1}^{n-1} E\big[(X_{i+1}-X_i)^2\big] \\
&= C \sum_{i=1}^{n-1} 2\sigma^2 \\
&= C \cdot 2(n-1)\sigma^2.
\end{aligned}
$$

要使 $E[\hat\sigma^2] = \sigma^2$，必须有
$$
C \cdot 2(n-1)\sigma^2 = \sigma^2.
$$

两边同时除以 $\sigma^2$（$\sigma^2>0$）得
$$
C \cdot 2(n-1) = 1
\quad\Rightarrow\quad
C = \frac{1}{2(n-1)}.
$$

### 练习题2：比较两个无偏估计量的有效性

> 设 $X_1, X_2, \dots, X_n$ 是来自同一总体的样本，满足  
> $$
> E[X_i] = \mu,\qquad D(X_i) = \sigma^2 < \infty,
> $$
> 且相互独立。  
> 定义两个用来估计 $\mu$ 的统计量：
> $$
> T_1 = \bar X = \frac{1}{n}\sum_{i=1}^n X_i,\qquad
> T_2 = X_1.
> $$
> 1. 证明 $T_1$ 和 $T_2$ 都是 $\mu$ 的无偏估计量。  
> 2. 分别计算 $D(T_1)$ 和 $D(T_2)$。  
> 3. 根据方差比较这两个估计量的有效性。

---

### 解答

#### 1. 无偏性

**(1) 对 $T_1=\bar X$：**

$$
E[T_1]
= E\left(\frac{1}{n}\sum_{i=1}^n X_i\right)
= \frac{1}{n}\sum_{i=1}^n E[X_i]
= \frac{1}{n}\cdot n\mu
= \mu.
$$

所以 $T_1$ 是 $\mu$ 的无偏估计量。

**(2) 对 $T_2=X_1$：**

$$
E[T_2] = E[X_1] = \mu.
$$

所以 $T_2$ 也是 $\mu$ 的无偏估计量。

---

#### 2. 方差计算

**(1) $T_1=\bar X$ 的方差**

因为 $X_i$ 独立同分布、方差为 $\sigma^2$，有

$$
D(T_1)
= D\left(\frac{1}{n}\sum_{i=1}^n X_i\right)
= \frac{1}{n^2}\sum_{i=1}^n D(X_i)
= \frac{1}{n^2}\cdot n\sigma^2
= \frac{\sigma^2}{n}.
$$

**(2) $T_2=X_1$ 的方差**

$$
D(T_2) = D(X_1) = \sigma^2.
$$

---

#### 3. 有效性比较

当 $n \ge 2$ 时，
$$
D(T_1) = \frac{\sigma^2}{n} < \sigma^2 = D(T_2).
$$

也就是说，在同样都是**无偏估计量**的前提下，$T_1$ 的方差更小，更“集中”，因此更有效。

**结论：**

- $T_1=\bar X$ 和 $T_2=X_1$ 都是 $\mu$ 的无偏估计量；
- 但 $T_1$ 的方差更小，是对 $\mu$ 更**有效**的估计量。

这道题说明了一个重要直觉：  
> 在估计同一个参数时，“利用全部样本信息”的估计量（样本均值）通常比“只用一个样本”的估计量（$X_1$）更有效。
