### 参数估计的基本框架

**背景：**  
有一组样本  
$$
X_1, X_2, \dots, X_n
$$  
来自某个带参数的总体分布（例如 $N(\mu,\sigma^2)$、$\text{Bernoulli}(p)$ 等）。  
我们希望根据样本，去“推回去”总体的一些**数字特征**（如均值 $E[X]$、方差 $D[X]$，或更一般的参数 $\theta$）。  
这个用样本信息去推断总体未知量的过程，就叫做**估计**。

直观理解：**“用局部（样本）去估计整体（总体参数）”。**

---

### 估计量 vs 估计值

- **估计量（estimator）**：  
  是样本的函数，是一个随机变量(大写)。  
  记为  
  $$
  \hat\theta = T(X_1,\dots,X_n)
  $$
  例如样本均值 $\bar X$、样本方差 $S^2$ 等。

- **估计值（estimate）**：  
  当样本具体取到数值 $x_1,\dots,x_n$ 时，带入估计量得到的具体数 (小写)。  
  例如  
  $$
  \bar x = \frac{1}{n}\sum_{i=1}^n x_i.
  $$

---

### 点估计与区间估计

- **点估计**：  
  给出一个**单一数值**作为总体参数的估计。  
  例如用 $\bar X$ 估计 $\mu$，用 $S^2$ 估计 $\sigma^2$。

- **区间估计**：  
  给出一个**区间**来估计参数，例如置信区间  
  $$
  [L(X_1,\dots,X_n),\, U(X_1,\dots,X_n)]
  $$
  满足
  $$
  P\{\theta \in [L,U]\} = 1-\alpha.
  $$  
  这里 $1-\alpha$ 就是**置信度**（在“较大概率”意义下覆盖真值）。

本节主要讨论的是**点估计**，特别是矩估计。

---

### 点估计的常见方法

常见两大类：

1. **矩估计（method of moments）**
2. **极大似然估计（maximum likelihood estimation, MLE）**

---

### 矩：原点矩与中心矩

对一个随机变量 $X$，其矩大致分两类：

- **原点矩（about 0）**：  
  $$
  \text{一阶原点矩： } E[X],\quad
  \text{二阶原点矩： } E[X^2],\quad
  \text{三阶原点矩： } E[X^3], \dots
  $$

- **中心矩（about 均值）**：  
  $$
  \text{一阶中心矩： } E[X - E[X]], \\
  \text{二阶中心矩： } E\big[(X - E[X])^2\big], \\
  \text{三阶中心矩： } E\big[(X - E[X])^3\big], \dots
  $$

在实际应用中，用得最多的是：

- 一阶原点矩：$E[X]$ —— **期望**  
- 二阶中心矩：$E\big[(X-E[X])^2\big]$ —— **方差**

---

### 矩估计的基本思想

**核心想法：**  
> 用“样本矩”去逼近“总体矩”，再通过“总体矩关于参数的表达式”解出参数的估计。

具体做法（原点矩为例）：

1. 写出总体的前若干个矩（用参数表示），例如  
   $$
   E[X] = m_1(\theta),\quad E[X^2] = m_2(\theta),\dots
   $$
2. 写出对应的**样本矩**：
   $$
   \text{样本一阶原点矩： } \bar X = \frac{1}{n}\sum_{i=1}^n X_i,
   $$
   $$
   \text{样本二阶原点矩： } \frac{1}{n}\sum_{i=1}^n X_i^2,\ \text{等等。}
   $$
3. 用“样本矩 ≈ 总体矩”建立方程，通常是直接令它们相等，例如  
   $$
   \bar X = m_1(\theta), \quad
   \frac{1}{n}\sum_{i=1}^n X_i^2 = m_2(\theta),
   $$
   解这个方程（组），得到参数的矩估计
   $$
   \hat\theta_{\text{MM}}.
   $$

---

### 矩估计的典型例子（均值与方差）

- 用**样本均值**估计总体均值：  
  $$
  \hat\mu = \bar X = \frac{1}{n}\sum_{i=1}^n X_i.
  $$  
  因为
  $$
  E[\bar X] = \mu,
  $$
  所以 $\bar X$ 是 $\mu$ 的**无偏估计量**。

- 用**样本方差**估计总体方差：  
  样本方差定义为  
  $$
  S^2 = \frac{1}{n-1}\sum_{i=1}^n (X_i - \bar X)^2,
  $$  
  可以证明
  $$
  E[S^2] = \sigma^2,
  $$
  所以 $S^2$ 是 $\sigma^2$ 的**无偏估计量**。

这两个估计量在很多教材中会同时被看作“矩估计”的自然结果（因为它们是由样本矩推导出来的），并且在正态总体下还有很多良好性质。

---

### 如何评价一个估计量的好坏？

常见的三条标准（点估计）：

1. **无偏性（unbiasedness）**  
   $$
   E[\hat\theta] = \theta.
   $$
   估计量的平均值等于真参数，即“长期平均不偏高也不偏低”。

2. **有效性（efficiency）**  
   在所有无偏估计量中，若某个估计量的方差最小，则称它是“最有效”的。  
   简单理解：波动越小越好。

3. **相合性（consistency）**  
   当样本量 $n\to\infty$ 时，估计量以概率 $1$ “逼近”真值：  
   $$
   \hat\theta \xrightarrow{P} \theta.
   $$
   简单理解：样本越多，估计越准、离真值越近。

---

### 小结

- **估计**：用样本信息去推断总体参数（均值、方差等）。  
- **估计量 / 估计值**：随机函数 vs 具体数值。  
- **点估计**：矩估计、极大似然估计；  
  **区间估计**：给出带置信度的参数区间。  
- **矩估计**：用样本矩替代表达参数的总体矩，建立方程解出参数。  
- 实用中最常见的矩估计：
  - 用 $\bar X$ 估计 $\mu$，因为 $E[\bar X]=\mu$；  
  - 用 $S^2$ 估计 $\sigma^2$，因为 $E[S^2]=\sigma^2$。  
- 评价估计量优劣的主要标准：**无偏性、有效性、相合性**。

### 练习题：矩估计

> 练习：设 $X_1, X_2, \dots, X_n$ 是来自某总体的样本，其概率密度函数为  
> $$
> f(x;\theta) =
> \begin{cases}
> (\theta+1)x^\theta, & 0<x<1, \\
> 0, & \text{其他},
> \end{cases}
> \qquad \theta>-1.
> $$
> 用矩估计求参数 $\theta$ 的估计量 $\hat\theta_{\text{MM}}$。

---

### 解答

**第 1 步：计算总体的一阶原点矩 $E[X]$**

$$
E[X]
= \int_0^1 x f(x;\theta)\,\mathrm{d}x
= \int_0^1 x (\theta+1)x^\theta \,\mathrm{d}x
= (\theta+1)\int_0^1 x^{\theta+1}\,\mathrm{d}x
= \frac{\theta+1}{\theta+2}\left.x^{\theta+2}\right|_0^1
= \frac{\theta+1}{\theta+2}
$$

记总体均值为  
$$
\mu = E[X] = \frac{\theta+1}{\theta+2}.
$$

**第 2 步：令样本矩等于总体矩**

样本一阶原点矩为样本均值：
$$
\bar X = \frac{1}{n}\sum_{i=1}^n X_i.
$$

矩估计法的基本思想是令
$$
\bar X \approx E[X].
$$

于是建立方程：
$$
\bar X = \frac{\theta+1}{\theta+2}.
$$


**第 3 步：解出 $\theta$ 的矩估计**

从
$$
\bar X = \frac{\theta+1}{\theta+2}
$$
出发，得

$$
\theta = \frac{2\bar X - 1}{1 - \bar X}.
$$

于是，参数 $\theta$ 的**矩估计量**为
$$
\hat\theta_{\text{MM}}
= \frac{2\bar X - 1}{1 - \bar X}
$$

> **说明：**
> * $\hat\theta$ 是一个**估计量**，它是一个关于样本 $\bar{X}$ 的函数（也是一个随机变量）。
> * 如果给定一组具体的样本观测值 $x_1, \dots, x_n$，我们可以计算出样本均值 $\bar{x}$，将其代入上式得到的**估计值** $\hat\theta_{\text{value}} = \frac{2\bar x - 1}{1 - \bar x}$。
> * 除了一阶矩（样本均值），也可以用二阶矩（如样本方差）构造别的矩估计，不过本题中用一阶矩已经足够唯一确定 $\theta$。而且，我们优先使用尽可能低阶的矩。
