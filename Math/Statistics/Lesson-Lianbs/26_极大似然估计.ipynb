### 极大似然估计（Maximum Likelihood Estimation, MLE）

#### 1. 基本思想

设  
- 样本 $X_1,X_2,\dots,X_n$ 独立同分布，来自某个带参数的分布 $f(x;\theta)$；  
- $\theta$ 是未知参数（可以是一维或多维）。

**极大似然估计的核心想法：**

> 在所有可能的参数取值中，选取那个使“已经观察到的这组样本”**出现的概率（或密度）最大**的参数值，作为对真参数的估计。

---

#### 2. 似然函数与对数似然

给定样本值 $x_1,\dots,x_n$，定义**似然函数**：

- 连续型：
  $$
  L(\theta)
  = L(\theta\mid x_1,\dots,x_n)
  = \prod_{i=1}^n f(x_i;\theta).
  $$
- 离散型：
  $$
  L(\theta)
  = \prod_{i=1}^n P_\theta(X_i = x_i).
  $$

由于 $\ln(\cdot)$ 是单调递增函数，最大化 $L(\theta)$ 等价于最大化**对数似然函数**：

$$
\ell(\theta)
= \ln L(\theta)
= \sum_{i=1}^n \ln f(x_i;\theta).
$$

在实际计算中，我们几乎总是用 $\ell(\theta)$ 而不是 $L(\theta)$。

---

#### 3. 一般求解步骤（单参数情形）

1. 写出似然函数：
   $$
   L(\theta) = \prod_{i=1}^n f(x_i;\theta).
   $$

2. 取对数得到对数似然：
   $$
   \ell(\theta) = \ln L(\theta).
   $$

3. 求导并令其为零（必要条件）：
   $$
   \frac{\mathrm{d}}{\mathrm{d}\theta}\ell(\theta) = 0.
   $$

4. 解出候选解 $\hat\theta$，再检查它确实使 $\ell(\theta)$（或 $L(\theta)$）取得最大值  
   （例如用二阶导数 $\ell''(\theta)<0$，或直接比较边界值）。

5. 得到的 $\hat\theta$ 就是参数的**极大似然估计量**。

---

#### 4. 估计量与估计值

- 对于随机样本 $X_1,\dots,X_n$ 而言，求得的
  $$
  \hat\theta_{\text{MLE}} = \hat\theta_{\text{MLE}}(X_1,\dots,X_n)
  $$
  是一个**估计量**（随机变量）。
- 当样本取到具体数值 $x_1,\dots,x_n$，带入上式：
  $$
  \hat\theta_{\text{MLE}}(x_1,\dots,x_n)
  $$
  就是具体的**估计值**。

---

#### 5. 与矩估计的对比（逻辑上如何理解）

- **矩估计**：  
  用“样本矩”去逼近“总体矩”，从而解出参数。  
  思路偏“匹配整体特征”。

- **极大似然估计**：  
  直接问：“在所有可能的 $\theta$ 中，哪一个最能解释我们已经观察到的这组数据？”  
  思路偏“解释能力最强”。

在不少常见模型（如正态、二项、泊松等）中，矩估计与 MLE 的形式往往比较接近，甚至相同（比如伯努利分布中都是样本均值）。

---

### 一个简单例子：伯努利分布的极大似然估计

设  
- $X_1,\dots,X_n \sim \text{Bernoulli}(p)$ 独立同分布，  
- 即 $P(X_i=1)=p,\ P(X_i=0)=1-p$，参数 $p\in(0,1)$。

**目标：** 用 MLE 估计 $p$。

---

#### 1. 写出似然函数

若观测到的样本为 $x_1,\dots,x_n\in\{0,1\}$，则

$$
L(p)
= \prod_{i=1}^n p^{x_i}(1-p)^{1-x_i}
= p^{\sum x_i}(1-p)^{n-\sum x_i}.
$$

记
$$
S = \sum_{i=1}^n x_i
$$
为样本中“1”的个数。

---

#### 2. 对数似然

$$
\ell(p)
= \ln L(p)
= S\ln p + (n-S)\ln(1-p).
$$

---

#### 3. 求导并令零

对 $p$ 求导：

$$
\frac{\mathrm{d}}{\mathrm{d}p}\ell(p)
= \frac{S}{p} - \frac{n-S}{1-p}.
$$

令其为零：

$$
\frac{S}{p} - \frac{n-S}{1-p} = 0
\quad\Rightarrow\quad
S(1-p) = p(n-S)
\quad\Rightarrow\quad
Sp = S
\quad\Rightarrow\quad
p = \frac{S}{n}.
$$

并且可验证这是极大值。

---

#### 4. 得到 MLE

伯努利参数 $p$ 的极大似然估计量为

$$
\hat p_{\text{MLE}}
= \frac{1}{n}\sum_{i=1}^n X_i
= \bar X,
$$

即**样本均值**。

这和矩估计用样本均值去估计 $E[X]=p$ 的结果完全一致，是一个典型的“矩估计与 MLE 一致”的例子。

---

### 小结（极大似然估计的几个要点）

1. **思想**：选使观测样本“最有可能发生”的参数值。  
2. **工具**：似然函数 $L(\theta)$ 与对数似然 $\ell(\theta)$。  
3. **步骤**：写 $L(\theta)$ → 写 $\ell(\theta)$ → 求导令零 → 解出 $\hat\theta$ → 检查为极大值。  

### 练习题：同时使用矩估计和极大似然估计

> 设 $X_1, X_2, \dots, X_n$ 是来自指数分布的样本，其概率密度函数为  
> $$
> f(x;\lambda) =
> \begin{cases}
> \lambda e^{-\lambda x}, & x>0, \\
> 0, & \text{其他},
> \end{cases}
> \qquad \lambda>0.
> $$
> 1. 用**矩估计法**求参数 $\lambda$ 的估计量 $\hat\lambda_{\text{MM}}$。  
> 2. 用**极大似然估计法**求参数 $\lambda$ 的估计量 $\hat\lambda_{\text{MLE}}$。  
> 3. 比较这两个估计量。

---

### 解答

#### 一、矩估计法

**第 1 步：计算总体一阶原点矩**

指数分布 $X\sim\text{Exp}(\lambda)$ 的期望为
$$
E[X] = \frac{1}{\lambda}.
$$

记总体均值为
$$
\mu = E[X] = \frac{1}{\lambda}.
$$

**第 2 步：令样本矩等于总体矩**

样本一阶原点矩为样本均值
$$
\bar X = \frac{1}{n}\sum_{i=1}^n X_i.
$$

矩估计思想：令
$$
\bar X = E[X] = \frac{1}{\lambda}.
$$

**第 3 步：解出 $\lambda$ 的矩估计**

由
$$
\bar X = \frac{1}{\lambda}
$$
得到
$$
\hat\lambda_{\text{MM}} = \frac{1}{\bar X}.
$$

---

#### 二、极大似然估计法

**第 1 步：写出似然函数**

对给定样本 $x_1,\dots,x_n$，似然函数为
$$
L(\lambda)
= \prod_{i=1}^n \lambda e^{-\lambda x_i}
= \lambda^n \exp\!\left(-\lambda\sum_{i=1}^n x_i\right).
$$

**第 2 步：对数似然函数**

$$
\ell(\lambda)
= \ln L(\lambda)
= n\ln\lambda - \lambda\sum_{i=1}^n x_i.
$$

**第 3 步：求导并令零**

$$
\frac{\mathrm{d}}{\mathrm{d}\lambda}\ell(\lambda)
= \frac{n}{\lambda} - \sum_{i=1}^n x_i.
$$

令其为零：
$$
\frac{n}{\lambda} - \sum_{i=1}^n x_i = 0
\quad\Rightarrow\quad
\lambda = \frac{n}{\sum_{i=1}^n x_i}.
$$

注意
$$
\bar X = \frac{1}{n}\sum_{i=1}^n x_i
\quad\Rightarrow\quad
\sum_{i=1}^n x_i = n\bar X,
$$
因此
$$
\hat\lambda_{\text{MLE}}
= \frac{n}{n\bar X}
= \frac{1}{\bar X}.
$$

检查二阶导数：
$$
\frac{\mathrm{d}^2}{\mathrm{d}\lambda^2}\ell(\lambda)
= -\frac{n}{\lambda^2} < 0,
$$
表明这是极大值点。

---

#### 三、比较矩估计与极大似然估计

- 矩估计得到：
  $$
  \hat\lambda_{\text{MM}} = \frac{1}{\bar X};
  $$
- 极大似然估计得到：
  $$
  \hat\lambda_{\text{MLE}} = \frac{1}{\bar X}.
  $$

**结论：**  
对于指数分布参数 $\lambda$，矩估计和极大似然估计给出的估计量完全相同，都是
$$
\boxed{\hat\lambda = \dfrac{1}{\bar X}}.
$$

这是一个很好用来对比两种方法思路、但结果一致的练习。
